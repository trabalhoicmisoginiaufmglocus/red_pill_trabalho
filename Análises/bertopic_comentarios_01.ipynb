```python
top_25_canais = df[1:26] #montando um novo dataframe com apenas as 25 primeiras linhas MENOS a linha zero
#cal. da porcentagem entre os 25 primeiros em relaÃ§Ã£o. ao total:
total_top_25 = top_25_canais['Count'].sum()
total_geral = df['Count'].sum()
porcentagem_top_25 = (total_top_25 / total_geral) * 100
print(f'{porcentagem_top_25:.2f}% do total.')
```
