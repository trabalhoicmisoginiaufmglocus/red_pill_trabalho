```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

folder_path = '/content/drive/MyDrive/red_pill_analises/'
caminho_canais = 'data_base/dados_limpos/canais_limpos.csv'
caminho_videos = 'data_base/dados_limpos/videos_limpos.csv'
caminho_comentarios = 'data_base/dados_limpos/comentarios_limpos.csv'

#df_canais_limpos = pd.read_csv(folder_path + caminho_canais)
df_videos_limpos = pd.read_csv(folder_path + caminho_videos)
df_comentarios_limpos = pd.read_csv(folder_path + caminho_comentarios)
```

## V√≠deos

```python
import pandas as pd
from collections import Counter
import numpy as np
import emoji # <-- IMPORTAMOS A NOVA BIBLIOTECA

# --- Prepara√ß√£o dos Dados ---
df_videos_limpos['emojis'] = df_videos_limpos['emojis'].fillna('')
total_de_videos = len(df_videos_limpos)

# ===================================================================
# PRIMEIRA ABORDAGEM (CORRIGIDA)
# ===================================================================
print("\n--- PRIMEIRA ABORDAGEM (CORRIGIDA COM BIBLIOTECA EMOJI) ---")

contador_presenca_emoji = Counter()

# Itera sobre cada v√≠deo que cont√©m emojis
for emoji_str in df_videos_limpos['emojis']:
    if pd.notna(emoji_str) and emoji_str:
        # üëá--- MUDAN√áA PRINCIPAL: Usamos a biblioteca para extrair emojis completos e √∫nicos ---üëá
        emojis_unicos_no_video = emoji.distinct_emoji_list(emoji_str)
        contador_presenca_emoji.update(emojis_unicos_no_video)

# Converte o contador para um DataFrame para melhor visualiza√ß√£o
df_presenca = pd.DataFrame(contador_presenca_emoji.items(), columns=['Emoji', 'Num_Videos'])
df_presenca['Percentual_Videos'] = (df_presenca['Num_Videos'] / total_de_videos) * 100
df_presenca = df_presenca.sort_values(by='Num_Videos', ascending=False)

# Salva o resultado limpo em um novo arquivo
df_presenca.to_csv(folder_path + 'data_base/resultados/emoji_contagem_por_texto_videos.csv', index=False, encoding='utf-8-sig')
print("\n‚úÖ Resultados corrigidos salvos com sucesso no arquivo 'emoji_contagem_por_texto_videos.csv'")

print("\n2. Pr√©-visualiza√ß√£o do resultado corrigido:")
display(df_presenca[:20])
```

### Mais profunda

```python
import pandas as pd
from collections import Counter
import numpy as np
import emoji
# ===================================================================
# SEGUNDA ABORDAGEM (CORRIGIDA COM BIBLIOTECA EMOJI)
# ===================================================================
print("\n--- SEGUNDA ABORDAGEM (An√°lise por Caractere Corrigida) ---")

# Lista para armazenar TODOS os emojis, incluindo repeti√ß√µes
lista_total_emojis = []

# Itera sobre cada c√©lula da coluna 'emojis'
for emoji_str in df_videos_limpos['emojis']:
    if pd.notna(emoji_str) and emoji_str:
        # üëá--- MUDAN√áA PRINCIPAL: Usamos emoji.emoji_list para pegar todos os emojis, n√£o apenas os √∫nicos ---üëá
        # A fun√ß√£o retorna uma lista de dicion√°rios, ent√£o extra√≠mos o valor da chave 'emoji'
        emojis_na_linha = [item['emoji'] for item in emoji.emoji_list(emoji_str)]
        lista_total_emojis.extend(emojis_na_linha)

# 1. Quantos caracteres, em toda a base, s√£o emojis?
# A contagem total agora √© simplesmente o tamanho da nossa lista final
total_caracteres_emojis = len(lista_total_emojis)

print(f"\n1. N√∫mero total de emojis (contando repeti√ß√µes e emojis complexos):")
print(f"   - Total: {total_caracteres_emojis} emojis")

# 2. Quais emojis aparecem mais? Qual a porcentagem?
# Usa o Counter na nossa lista completa de emojis
contador_frequencia_emoji = Counter(lista_total_emojis)

# Converte para DataFrame para visualiza√ß√£o
df_frequencia = pd.DataFrame(contador_frequencia_emoji.items(), columns=['Emoji', 'Frequencia_Total'])

# Calcula o percentual com base no novo total
if total_caracteres_emojis > 0:
    df_frequencia['Percentual_Total'] = (df_frequencia['Frequencia_Total'] / total_caracteres_emojis) * 100
else:
    df_frequencia['Percentual_Total'] = 0

df_frequencia = df_frequencia.sort_values(by='Frequencia_Total', ascending=False)

# Salva o resultado limpo em um novo arquivo
df_frequencia.to_csv(folder_path + 'data_base/resultados/emoji_contagem_frequencia_videos.csv', index=False, encoding='utf-8-sig')
print("\n‚úÖ Resultados de frequ√™ncia corrigidos salvos com sucesso no arquivo 'emoji_contagem_frequencia_videos.csv'")

print("\n2. Pr√©-visualiza√ß√£o da frequ√™ncia de cada emoji:")
display(df_frequencia[:10])
```
