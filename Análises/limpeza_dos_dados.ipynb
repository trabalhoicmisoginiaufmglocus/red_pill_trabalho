# üßº Limpeza dos dados

## ‚ú® Fun√ß√µes Gerais

Fun√ß√£o respons√°vel por criar uma nova coluna s√≥ com emojis (n√£o tira emojis da coluna text)

```python
def extract_emojis(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # Emojis de sorriso e emo√ß√£o
        "\U0001F300-\U0001F5FF"  # S√≠mbolos e pictogramas
        "\U0001F680-\U0001F6FF"  # Transporte e mapas
        "\U0001F700-\U0001F77F"  # S√≠mbolos alqu√≠micos
        "\U0001F780-\U0001F7FF"  # Geom√©tricos estendidos
        "\U0001F800-\U0001F8FF"  # S√≠mbolos suplementares A
        "\U0001F900-\U0001F9FF"  # S√≠mbolos suplementares B
        "\U0001FA00-\U0001FA6F"  # S√≠mbolos suplementares C
        "\U0001FA70-\U0001FAFF"  # S√≠mbolos suplementares D
        "\U00002702-\U000027B0"  # Dingbats
        "\U00002600-\U000026FF"  # Diversos s√≠mbolos
        "\U0000FE0E-\U0000FE0F"  # Varia√ß√£o de seletores
        "\U0001F1E0-\U0001F1FF"  # Bandeiras (Cantois regionais)
        "\U0000200D"             # Joiner de varia√ß√£o zero
        "]+", flags=re.UNICODE)
    emojis = emoji_pattern.findall(text)
    return "".join(emojis)
```

Fun√ß√£o respons√°vel por criar uma nova coluna s√≥ com as hashtags (n√£o tira hashtags da coluna text)

```python
def extract_hashtags(text):
    hashtag_pattern = re.compile(r"#\w+")
    hashtags = hashtag_pattern.findall(text)
    return " ".join(hashtags)
```

## üé¨ Canais

Concatenar as colunas '*tittle*' e '*description*' do DataFrame `df_canais_anonimizados` em uma nova coluna chamada '*text*'.

```python
df_canais_anonimizados['text'] = df_canais_anonimizados['description'].fillna('')
```

Remover quebras de linha `(\n)` e tabula√ß√µes `(\t)`;

```python
df_canais_anonimizados['text'] = df_canais_anonimizados['text'].str.replace(r'[\n\t]', '', regex=True)
```

Remover as men√ß√µes anonimizadas (afetam as an√°lises)

```python
df_canais_anonimizados['text'] = (
df_canais_anonimizados['text']
    .str.replace(r'\[MENCAO\]', ' ', regex=True)
    .str.replace(r'\[EMAIL\]', ' ', regex=True)
)
```

Retirar colunas '*tittle*' e '*description*'

```python
df_canais_anonimizados = df_canais_anonimizados.drop(columns=['description'], axis=1)
```

Criar uma nova coluna s√≥ com *emojis*

```python
df_canais_anonimizados['emojis'] = df_canais_anonimizados['text'].apply(extract_emojis)
```

Criar uma nova coluna s√≥ com as *hashtags*

```python
df_canais_anonimizados['hashtags'] = df_canais_anonimizados['text'].apply(extract_hashtags)
```

Criar um novo arquivo para *Canais Limpos*

```python
df_canais_anonimizados.to_csv(folder_path + 'data_base/dados_limpos/canais_limpos.csv', index=False)
```
