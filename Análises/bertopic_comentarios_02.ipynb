```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')
folder_path = '/content/drive/MyDrive/red_pill_analises/data_base/resultados/ResultadosComentarios/'

arq1 = 'topic_descriptions_final.csv'
df1 = pd.read_csv(folder_path + arq1, sep=';')
```

```python
from IPython.display import display
display(df1.drop(columns=['Representative_Docs']))
```

```python
top_25_canais = df1[1:15] #montando um novo dataframe com apenas as 25 primeiras linhas MENOS a linha zero
#cal. da porcentagem entre os 25 primeiros em rela√ß√£o. ao total:
total_top_25 = top_25_canais['Count'].sum()
total_geral = df1['Count'].sum()
porcentagem_top_25 = (total_top_25 / total_geral) * 100
print(f'{porcentagem_top_25:.2f}% do total.')
```

```python
#slicing [in√≠cio(inclusivo):fim(exclusivo)]
top_14_topicos = df1[1:15]

# 2. Encontre no DataFrame original as linhas que cont√™m os t√≥picos 17, 40 e 47
novas_linhas = df1[df1['Topic'].isin([17, 40, 47])]

# 3. Combine o DataFrame original e as novas linhas
# O ignore_index=True √© importante para criar um novo √≠ndice consecutivo
top_14_topicos_atualizado = pd.concat([top_14_topicos, novas_linhas], ignore_index=True)

# 4. Calcule a soma do 'Count' do novo DataFrame
total_top_25 = top_14_topicos_atualizado['Count'].sum()

# 5. Calcule a soma total do 'Count' do DataFrame original
total_geral = df1['Count'].sum()

# 6. Calcule a porcentagem
porcentagem_top_25 = (total_top_25 / total_geral) * 100

# Exiba o resultado
display(top_14_topicos_atualizado)
print('\n---')
print(f'A porcentagem do Count do novo DataFrame em rela√ß√£o ao total √©: {porcentagem_top_25:.2f}%')
```

```python
# remover t√≥picos indesejados
df_sem_indesejados = df1[~df1['Topic'].isin([2, -1])]

# pegar os 14 primeiros depois da remo√ß√£o
top_14_topicos = df_sem_indesejados.iloc[:14]

# definir novos t√≥picos e pegar s√≥ os que N√ÉO est√£o no top_14_topicos
novos_topics = [14, 17, 40, 47]
novas_linhas = df1[
    df1['Topic'].isin(novos_topics) &
    ~df1['Topic'].isin(top_14_topicos['Topic'])
]

# concatenar e resetar √≠ndice
topicos_atualizados = pd.concat([top_14_topicos, novas_linhas], ignore_index=True)
topicos_atualizados = topicos_atualizados.reset_index(drop=True)

# soma e porcentagem
total_topicos_atualizados = topicos_atualizados['Count'].sum()
total_geral = df1['Count'].sum()
porcentagem = (total_topicos_atualizados / total_geral) * 100

display(topicos_atualizados)
print(f'\nPorcentagem: {porcentagem:.2f}%')
```
