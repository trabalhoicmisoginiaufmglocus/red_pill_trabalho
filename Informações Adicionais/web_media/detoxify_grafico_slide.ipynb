```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Iniciando script para gerar os Gráficos de Dispersão (Versão 6, Bolinhas Maiores)...")

# =============================================================================
# 1. TABELAS DE DADOS
# =============================================================================

dados_videos = {
    'Canal': list(range(1, 29)),
    'Misoginia_Videos': [0,0,1,3,0,0,0,1,2,3,0,0,0,0,0,0,5,0,1,0,2,0,1,0,0,0,0,5],
    'Toxicidade_Videos': [0,2,3,8,1,1,1,2,8,7,0,0,1,0,1,0,16,0,3,0,3,0,2,11,1,0,3,9]
}

dados_comentarios = {
    'Canal': list(range(1, 29)),
    'Misoginia_Comentarios': [11,12,13,15,10,14,12,8,14,10,11,14,10,6,7,13,10,18,8,5,13,3,16,15,10,10,6,13],
    'Toxicidade_Comentarios': [16,16,18,21,15,18,16,11,19,14,15,18,15,9,11,18,15,23,11,6,18,5,21,21,15,15,10,23]
}

# =============================================================================
# 2. CRIAÇÃO DO DATAFRAME UNIFICADO
# =============================================================================

df_videos = pd.DataFrame(dados_videos)
df_comentarios = pd.DataFrame(dados_comentarios)

# Une os dois dataframes com base no Canal
df_merged = pd.merge(df_videos, df_comentarios, on='Canal').set_index('Canal')

# =============================================================================
# 3. GERAÇÃO DO SCATTER PLOT (TOXICIDADE) - VERSÃO SLIDE / BOLINHAS MAIORES
# =============================================================================

print("Gerando grafico_scatter_toxicidade_sem_grid_v6.png...")

sns.set_theme(style="white")
plt.figure(figsize=(15, 11))

ax1 = sns.scatterplot(
    data=df_merged,
    x='Toxicidade_Videos',
    y='Toxicidade_Comentarios',
    s=250,       # ✅ Tamanho das bolinhas aumentado para 250
    color='darkred',
    alpha=0.5
)

# --- Melhorias Visuais (Sem título, sem linha) ---
plt.xlabel("Percentual Toxicidade - Vídeos", fontsize=30, labelpad=30, fontweight='bold')
plt.ylabel("Percentual Toxicidade - Comentários", fontsize=30, labelpad=30, fontweight='bold')

plt.xticks(fontsize=18, fontweight='bold')
plt.yticks(fontsize=18, fontweight='bold')

plt.xlim(-1, 25)
plt.ylim(0, 25)

plt.xticks(range(0, 26, 5))
plt.yticks(range(0, 26, 5))

sns.despine()
ax1.spines['bottom'].set_linewidth(3)
ax1.spines['left'].set_linewidth(3)

plt.tight_layout()
plt.savefig('grafico_scatter_toxicidade_sem_grid_v6.png', dpi=400)
# plt.close() # Removido para exibir o gráfico

# =============================================================================
# 4. GERAÇÃO DO SCATTER PLOT (MISOGINIA) - VERSÃO SLIDE / BOLINHAS MAIORES
# =============================================================================

print("Gerando grafico_scatter_misoginia_sem_grid_v6.png...")

sns.set_theme(style="white")
plt.figure(figsize=(15, 11)) # Novo gráfico

ax2 = sns.scatterplot(
    data=df_merged,
    x='Misoginia_Videos',
    y='Misoginia_Comentarios',
    s=250,      # ✅ Tamanho das bolinhas aumentado para 250
    color='darkred',
    alpha=0.5
)

# --- Melhorias Visuais (Sem título, sem linha) ---
plt.xlabel("Percentual de Misoginia - Vídeos", fontsize=30, labelpad=30, fontweight='bold')
plt.ylabel("Percentual de Misoginia - Comentários", fontsize=30, labelpad=30, fontweight='bold')

plt.xticks(fontsize=18, fontweight='bold')
plt.yticks(fontsize=18, fontweight='bold')

plt.xlim(-1, 25)
plt.ylim(0, 25)

plt.xticks(range(0, 26, 5))
plt.yticks(range(0, 26, 5))

sns.despine()
ax2.spines['bottom'].set_linewidth(3)
ax2.spines['left'].set_linewidth(3)

plt.tight_layout()
plt.savefig('grafico_scatter_misoginia_sem_grid_v6.png', dpi=400)
# plt.close() # Removido para exibir o gráfico

print("\nScript finalizado. Arquivos salvos com sucesso.")
print("1. grafico_scatter_toxicidade_sem_grid_v6.png")
print("2. grafico_scatter_misoginia_sem_grid_v6.png")
```
